```python
import os
import numpy as np
import networkx as nx
from networkx import random_regular_graph, random_graphs

import pickle
```

```python
for N in range(6, 8,2):
    for r in range(1):
        print('\n\n (N, r): ', (N, r))
        instance_dir = './N_' + str(N)
        os.makedirs(instance_dir, exist_ok=True)

        G = random_regular_graph(3, N)
        coef_list = np.random.uniform(-1, 1, len(G.edges()) + N)
        coef_list = np.round(coef_list, 4)
        print(coef_list)
        print(len(coef_list))

        with open(instance_dir + '/QUBO_' + str(N) + 'V_3reg_'+ '.gpickle', 'wb') as f:
            pickle.dump(G, f, pickle.HIGHEST_PROTOCOL)

        np.savetxt(instance_dir + '/QUBO_coeff_' + str(N) + 'V_3reg_'+ 'r_'+ str(r)+ '.txt', coef_list)
```

```python
for N in range(6, 12, 2):
    for r in range(2):
        print('\n\n (N, r): ', (N, r))
        instance_dir = './N_' + str(N)
        os.makedirs(instance_dir, exist_ok=True)

        # Generate a complete graph with N nodes
        G = nx.complete_graph(N)
        
        # Coefficients for edges and nodes
        coef_list = np.random.uniform(-1, 1, len(G.edges()) + N)
        coef_list = np.round(coef_list, 4)
        print(coef_list)
        # print(len(coef_list))

        # Save the graph
        with open(instance_dir + '/QUBO_' + str(N) + 'V_comp_' + '.gpickle', 'wb') as f:
            pickle.dump(G, f, pickle.HIGHEST_PROTOCOL)

        # Save the coefficients
        np.savetxt(instance_dir + '/QUBO_coeff_' + str(N) + 'V_comp_' + 'r_' + str(r) + '.txt', coef_list)
```

```python
for N in range(6, 12, 2):
    for r in range(2):
        print('\n\n (N, r): ', (N, r))

        instance_dir = './N_' + str(N)
        os.makedirs(instance_dir, exist_ok=True)

        # Generate a complete graph with N nodes
        G = nx.complete_graph(N)
        
        # Coefficients for edges and nodes
        coef_list = np.random.uniform(-1, 1, len(G.edges()) + N)
        coef_list = np.round(coef_list, 4)
        print(coef_list)
        # print(len(coef_list))

        # Save the graph
        with open(instance_dir + '/QUBO_' + str(N) + 'V_comp_' + '.gpickle', 'wb') as f:
            pickle.dump(G, f, pickle.HIGHEST_PROTOCOL)

        # Save the coefficients
        np.savetxt(instance_dir + '/QUBO_coeff_' + str(N) + 'V_comp_' + 'r_' + str(r) + '.txt', coef_list)
        
        ################### read ##########


        #region load qubo instances, get Hamiltonian and edge_coeff_dict
        instance_dir = './N_' + str(N)
        with open(instance_dir + '/QUBO_' + str(N ) + 'V_comp_'+ '.gpickle', 'rb') as f:
            G = pickle.load(f)
        edge_list = G.edges()
        coeff_list = np.loadtxt(instance_dir + '/QUBO_coeff_' + str(N) + 'V_comp_'+ 'r_'+ str(r)+ '.txt')
        h_list = coeff_list[: N ]
        J_list = coeff_list[N:]

        # Initialize dictionary for single Pauli Z term with coefficient from h_list
        edge_coeff_dict = {}
        edge_coeff_dict.update({(i,): h_val for i, h_val in enumerate(h_list)})
        #Initialize dictionary for Pauli ZZ term with coefficient from from J_list
        edge_coeff_dict.update({edge: J_val for edge, J_val in zip(edge_list, J_list)})
        print('edge_coeff_dict' , edge_coeff_dict)
        print(len(edge_coeff_dict.values()))
```
